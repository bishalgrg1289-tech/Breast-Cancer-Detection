```python
import numpy as np
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator # type: ignore
training_data_generator=ImageDataGenerator(
    rescale=1./255,
    horizontal_flip=True,
    vertical_flip=True,
    shear_range=0.2,
    zoom_range=0.2
)
training_data=training_data_generator.flow_from_directory(
    'Data_set/training_set',
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical',
    color_mode='grayscale'
)
```

```python
test_data_generator=ImageDataGenerator(
    rescale=1./255
)
test_set=test_data_generator.flow_from_directory(
    'Data_set/test_set',
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical',
    color_mode='grayscale'
)
```

```python
nn=tf.keras.models.Sequential()
```

```python
nn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[128, 128, 1]))
nn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
nn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
nn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
nn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
nn.add(tf.keras.layers.Flatten())
```

```python
nn.add(tf.keras.layers.Dense(units=128, activation='relu'))
nn.add(tf.keras.layers.Dense(units=3, activation='softmax'))
```

```python
nn.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
nn.fit(x=training_data,validation_data=test_set,epochs=20)
```

```python
from tensorflow.keras.preprocessing import image # type: ignore
test_image = image.load_img('Data_set/single_prediction/benign_1.png', target_size = (128, 128),color_mode='grayscale')
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
test_image = test_image / 255.0
result = nn.predict(test_image)
print(result)
training_data.class_indices
if result[0][0]>result[0][1] and result[0][0]>result[0][2]:
    print('benign')
elif result[0][1]>result[0][0] and result[0][1]>result[0][2]:
    print('malignant')
else:
    print('normal')
```
